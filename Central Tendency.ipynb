```python
import numpy as np
from scipy import stats
```

```python
# Sample dataset
data = [12, 15, 12, 18, 19, 20, 21, 20, 20, 25, 100]  # Notice the outlier (100)
```

```python
# Mean
mean = np.mean(data)
```

```python
# Median
median = np.median(data)
```

```python
# Mode
mode_result = stats.mode(data, keepdims=True)
mode = mode_result.mode[0]
mode_count = mode_result.count[0]
```

```python
# Output results
print("Dataset:", data)
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode} (appears {mode_count} times)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=10, kde=True, color='lightblue', edgecolor='black')
```

```python
# Add vertical lines for Mean, Median, Mode
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean:.2f}')
plt.axvline(median, color='green', linestyle='-.', linewidth=2, label=f'Median = {median}')
plt.axvline(mode, color='blue', linestyle=':', linewidth=2, label=f'Mode = {mode}')
```

```python
# Labels and Legend
plt.title('Measures of Central Tendency')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
