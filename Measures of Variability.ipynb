```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample dataset (includes an outlier)
data = [12, 15, 12, 18, 19, 20, 21, 20, 20, 25, 100]
```

```python
# Calculating measures of variability
data_range = np.max(data) - np.min(data)
variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)   # Sample standard deviation
mean = np.mean(data)
```

```python
# Print results
print("Dataset:", data)
print(f"Range: {data_range}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# Plotting the distribution
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=10, kde=True, color='skyblue', edgecolor='black')
```

```python
# Plot vertical lines and capture them for the legend
mean_line = plt.axvline(mean, color='red', linestyle='--', linewidth=2)
std_plus_line = plt.axvline(mean + std_dev, color='orange', linestyle=':', linewidth=2)
std_minus_line = plt.axvline(mean - std_dev, color='orange', linestyle=':', linewidth=2)
```

```python
# Add legend manually
plt.legend(
    handles=[mean_line, std_plus_line, std_minus_line],
    labels=[
        f'Mean = {mean:.2f}',
        f'+1 SD = {mean + std_dev:.2f}',
        f'-1 SD = {mean - std_dev:.2f}'
    ]
)
```

```python
# Add titles and labels
plt.title('Measures of Variability')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data = [10, 15, 14, 8, 12, 11, 13]
```

```python
# Step 1: Calculate the mean
mean = sum(data) / len(data)

# Step 2: Calculate absolute deviations and MAD
abs_deviations = [abs(x - mean) for x in data]
mad = sum(abs_deviations) / len(data)

# Output
print("Data:", data)
print("Mean:", mean)
print("Absolute Deviations:", abs_deviations)
print("Mean Absolute Deviation (MAD):", mad)
```

```python
n = len(data)

# Step 1: Calculate the mean
mean = sum(data) / n
print("Mean:", mean)

# Step 2: Compute squared deviations
squared_devs = [(x - mean) ** 2 for x in data]
```

```python
# Step 3: Compute variance (population)
variance = sum(squared_devs) / n
print("Variance:", variance)

# Step 4: Standard deviation is the square root of variance
std_dev = variance ** 0.5
print("Standard Deviation:", std_dev)
```

```python
# ✅ Compare with NumPy (for verification)
np_variance = np.var(data)            # population variance
np_std_dev = np.std(data)             # population standard deviation

print("\nNumPy Calculation:")
print("Variance (NumPy):", np_variance)
print("Standard Deviation (NumPy):", np_std_dev)
```
